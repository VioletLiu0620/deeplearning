### One of the msot common errors in deep learnings: shape error

(# of rows, # of coloumns)

1. The **inner dimension** must match:

* `(3, 2) @ (3, 2)` won't work
* `(2, 3) @ (3, 2)` will work

2. The resulting matrix has the shape of the **outer dimensions**:

* `(2, 3) @ (3, 2)` will work, and result on a shape of 2, 2

```python
tensor_A = torch.tensor([[1, 2],
                        [3, 4],
                        [5, 6]])

tensor_B = torch.tensor([[7, 10],
                        [8, 11],
                        [9, 12]])

# we can use the abbreviation of torch.mm instead of torch.matmul
torch.mm(tensor_A, tensor_B)
```

```python
# use transpost with .T
tensor_B, tensor_B.T, tensor_B.T.shape
```

## Reshaping, stacking, squeezing and unsequeezing tensors

* Reshaping - reshaping an input tensor to a defined shape
* View - Return a view of a input tensor of certain shape but keep the same memory as the original tensor
* Stacking - combine mutiple tensors on top of each other  (vstack) or side b side (hstack)
* Squeeze - remove all `i` dimensions from a tensor
* Unsqueeze - add a `i` dimensiosn to a tensor
* Permute - Return a view of the input with dimensions permuted (swapped) in a certain way

```python
import torch
x = torch.arange(1. ,10.)
x, x.shape
```

```python
# the reshape has to be compatible, adding new square bracket is ok, but the # of element in the original tensor must be able to fit in the new size, 1x10 to 5x2 is ok, but no creation of new element

x_reshaped = x.reshape(9,1,1)
x_reshaped, x_reshaped.shape
```

```python
# change the view
z = x_reshaped.view(9,1)
z, z.shape

## Changing z changes x (because a view of a tensor shares the same memory as the original tensor)
```

```python
z[:, 0 ] = 5
z, x_reshaped
```

```python
# Stack tensors on top of each other
x
x_stacked = torch.stack([x, x, x])
x_stacked
```

```python
# torch.squeeze() - removes all single dimensions from a target tensor
y = torch.arange(0,10,dtype= torch.float32)
y = y.reshape(1,10)
y
```

```python
y.shape
```

```python
y.squeeze()
```

```python
y.squeeze().shape
```

```python
# torch.unsqueeze() - adds a single dimension to a target tensor at a specfic dim
print(f"Previous target: {y.squeeze()}")
print(f"Previous shape: {y.squeeze().shape}")

x_unsqueezed = y.squeeze().unsqueeze(dim=0)
print(f"New target: {x_unsqueezed}")
print(f"New shape: {x_unsqueezed.shape}")
```

```python
# torch.permute - rearranges the dimensions of a target tensor in a specified order

x_original = torch.rand(size=(224,224,3)) # height, width, colour_channels

# permute the original tensor to rearrange the axis or dim order
# switch the colour channel to the first

x_permuted = x_original.permute(2, 0 ,1) # shift the order 0->1, 1->2, 2->0

print(f"Previous shape: {x_original.shape}")
print(f"New shape: {x_permuted.shape}")
```

## PyTorch tensors & Numpy

NumPy is a popular scientific python numerical computing library.

* Data in NumPy, want in PyTorch tensor -> `torch.from_numpy(ndarray)`
* PyTorch tensor -> Numpy -> `torch.Tensor.numpy()`

```python
# Numpy array to tensor
import torch
import numpy as np

array = np.arange(1.0, 8.0)
tensor = torch.from_numpy(array)

# warning, when converting from numpy -> pytorch, pytorch reflects numpy's default datatype of float64 unless specified

array, tensor
```

```python
array.dtype
```

```python
tensor.dtype
```

```python
# Change the value of array
array = array + 1
array, tensor

# the new tensor didn't change, when use from_numpy, create one new in the memory
```

```python
# Tensor to Numpy array
tensor = torch.ones(7)
numpy_tensor = tensor.numpy()
tensor, numpy_tensor
```

```python
# Change the tensor, what happen to NumPy tensor?
tensor = tensor + 1
tensor, numpy_tensor

# the new tensor didn't change, don't share memory
```

## Reproducbility (trying to take random out of random)

In short, how a nerual network learns:

`start with random numbers -> tensor operations -> update random numbers to try and make them better represenation of the data -> again -> again -> again...`

to reduce the randomness in neural netwroks and PyTorch comes the conept of a **random seed**

what the random seed does is "flavour" the randomness

```python
import torch

# Create two random tensors
random_tensor_A = torch.rand(3,4)
random_tensor_B = torch.rand(3,4)
random_tensor_A, random_tensor_B
```

```python
print(random_tensor_B == random_tensor_A)
```

```python
# Let's make some random but reproducible tensors
import torch

# Set the random seed
RANDOM_SEED = 42
# answer to the univers because it sounded funny??
# random seed: you get the same value if random seed value is the same
# be able to reproduce the same tensor even when refreshing the page
# for example, random_seed = 42 -> [0][0][0] = 0.8823 as always

torch.manual_seed(RANDOM_SEED)
random_tensor_C = torch.rand(3, 4)

torch.manual_seed(RANDOM_SEED)
random_tensor_D = torch.rand(3, 4)

print(random_tensor_C)
print(random_tensor_D)
print(random_tensor_C == random_tensor_D)
```
